```python
var("t","s")
f(t,s)=1-(s*cos(4*t)*cos(t))/sqrt(1-s^2*(cos(4*t))^2*(sin(t))^2)
```

```python
x(t)=f(t-pi/2,s)
y(t)=f(t,s)
```

```python
parametric_plot([x(t).subs(s=0.5),y(t).subs(s=0.5)],(t,0,2*pi))
```

```python
plots_list=[]
colors=["#FF0000","#FFB900","#F7FF00","#70FF00","#00FF9E","#00FFFF","#0074FF","#7000FF","#F300FF","#B200FF"]
colors.reverse()
for i in range(1,11):
    plots_list.append(parametric_plot([x(t).subs(s=i/10),y(t).subs(s=i/10)],(t,0,2*pi),color=colors[i-1]))
```

```python
plot_final=plots_list[0]
for i in range(1,10):
    plot_final+=plots_list[i]
```
