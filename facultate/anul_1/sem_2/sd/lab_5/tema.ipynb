# Driven oscillator



The equation for the undamped case reads
$$ \begin{align*}
x''(t) + \omega_0^2 x(t) = F \cos(\omega t).
\end{align*} $$
One can get the solution with "variation of the constants" (twice). We can also use the Fourier transforms and obtain
$$ \begin{align*}
x(t) = 	K_1 \cos(\omega_0 t) + K_2 \sin(\omega_0 t) - F\frac{\cos(\omega t)}{\omega^2 - \omega_0^2}.
\end{align*} $$


*Proof.* let us make the following ansatz for the particular solution, $x_p(t) = \int_{-\infty}^{+\infty} X(k) \mathrm e^{-2\pi \mathrm i  kt} \, \mathrm dk$, where we have used the inverse Fourier transform. This means that
$$ \begin{align*}
\frac{\mathrm d^2}{\mathrm dt^2} x(t) &= \frac{\mathrm d^2}{\mathrm dt^2} \int_{-\infty}^{+\infty} X(k) \mathrm e^{-2\pi \mathrm i  kt} \, \mathrm dk \\
&=  \int_{-\infty}^{+\infty} X(k) \left( \frac{\mathrm d^2}{\mathrm dt^2} \mathrm e^{-2\pi \mathrm i  kt} \right) \, \mathrm dk \\
&= (-2\pi \mathrm i)^2 \int_{-\infty}^{+\infty} X(k) \mathrm e^{-2\pi \mathrm i  kt} \, \mathrm dk  
\end{align*} $$
We also need to know the inverse transform of the right hand side. Thus first,
$$ \begin{align*}
\mathcal F_k[\cos(\omega t)] &= \int_{-\infty}^{+\infty} \cos(\omega t) \mathrm e^{2\pi \mathrm i kt} \, \mathrm dt \\
&= \int_{-\infty}^{+\infty} \cos(2\pi \frac{\omega}{2\pi} t) \mathrm e^{2\pi \mathrm i kt} \, \mathrm dt \\
&= \frac{1}{2} ( \delta(k - \frac{\omega}{2\pi}) + \delta(k + \frac{\omega}{2\pi} ),
\end{align*} $$
hence we  have with the help of Dirac's "delta function"
$$ \begin{align*}
	\cos(\omega t) &= \mathcal F_t[ \frac{1}{2} ( \delta(k - \frac{\omega}{2\pi}) + \delta(k + \frac{\omega}{2\pi} ) ] \\
	&= \int_{-\infty}^{+\infty} \frac{1}{2} ( \delta(k - \frac{\omega}{2\pi}) + \delta(k + \frac{\omega}{2\pi} ) \mathrm e^{-2\pi \mathrm i kt} \, \mathrm dk
\end{align*} $$
Insert the inverse transforms into the  equation of motion to obtain
$$ \small{\begin{align*}
(-2\pi \mathrm i)^2 \int_{-\infty}^{+\infty} X(k) \mathrm e^{-2\pi \mathrm i  kt} dk + \omega_0^2 \int_{-\infty}^{+\infty} X(k) \mathrm e^{-2\pi \mathrm i  kt} \mathrm dk = 	\int_{-\infty}^{+\infty} \frac{1}{2} ( \delta(k - \frac{\omega}{2\pi}) + \delta(k + \frac{\omega}{2\pi} ) \mathrm e^{-2\pi \mathrm i kt} \, \mathrm dk
\end{align*}}$$
The equation is valid when the integrand match, thus
$$ \begin{align*}
	-(2\pi)^2 X(k) + \omega_0^2 X(k) = \frac{1}{2} ( \delta(k - \frac{\omega}{2\pi}) + \delta(k + \frac{\omega}{2\pi} ).
\end{align*} $$
We solve the equation for $X(k)$, and then use the inverse transform to get an answer for a particular solution.
$$ \begin{align*}
x_p(t) &= \int_{-\infty}^{+\infty} X(k) \mathrm e^{-2\pi \mathrm i  kt} \, \mathrm dk \\
&= \int_{-\infty}^{+\infty} \left(  \frac{1}{2} \frac{\delta(k - \frac{\omega}{2\pi}) + \delta(k + \frac{\omega}{2\pi})}{ \omega_0^2 - (2\pi)^2 k^2 } \right) \mathrm e^{-2\pi \mathrm i  kt} \, \mathrm dk \\
&= \frac{1}{2} \left(  \frac{\mathrm e^{-2\pi \mathrm i (\omega/(2\pi)) t}}{\omega_0^2 - (2\pi)^2 (\omega/2\pi)^2} + \frac{\mathrm e^{-2\pi \mathrm i (-\omega/(2\pi)) t}}{\omega_0^2 - (2\pi)^2 (-\omega/2\pi)^2}  \right) \\
&= \frac{1}{2} \frac{\mathrm e^{-\mathrm i\omega t} + \mathrm e^{+\mathrm i\omega t}}{\omega_0^2 - \omega^2} \\
&= \frac{\cos(\omega t)}{\omega_0^2 - \omega^2}.
\end{align*} $$


Se consideră ecuația diferențială ce descrie mișcarea oscilatorului armonic (fără frecare) asupra căaruia acționează o forță periodică de forma $F(t) = F_0 cos(\omega t)$:

$x''+\omega_0^2x=F_0cos(\omega t) $

a)Determinați soluția generală a ecuației în cazul $\omega_0 \neq \omega$ (cazul de nerezonanță);

-----------


Bineînțeles, mai întâi declarăm tot ce avem nevoie, variabile și funcții.

```python
var("t,F_0,omega,omega_0")
x=function("x")(t)
```

Acum declaram ecuația (și o afișăm ca să fim siguri că am scris-o cum trebuie)

```python
deq=diff(x,t,2)+omega_0^2*x==F_0*cos(omega*t)
show(deq)
```

Iar acum putem rezolva ecuația, după ce dăm assume-urile necesare ($\omega_0 \neq \omega$ din cerința subpunctului a), iar $\omega_0 \gt 0$ pentru ca desolve-ul să poată opera --- $\omega_0 \lt 0 $ nu ar avea sens deoarece pulsația nu poate fi negativă, iar $\omega = 0$ nu ar mai însemna mișcare, fapt ce nu ar fi interesant) 

```python
assume(omega_0>0)
assume(omega_0!=omega)
dsol=desolve(deq,[x,t])
show(dsol)
```
